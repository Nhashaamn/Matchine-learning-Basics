Scikit-learn provides several imputers to handle missing values in datasets. Here are the main ones:

1. **SimpleImputer**:
   - Fills missing values with a specified strategy (e.g., mean, median, most frequent, or constant value).
   ```python
   from sklearn.impute import SimpleImputer
   ```

2. **KNNImputer**:
   - Fills missing values using the k-Nearest Neighbors algorithm.
   ```python
   from sklearn.impute import KNNImputer
   ```

3. **IterativeImputer**:
   - Fills missing values using a multivariate imputation method, often referred to as chained equations or multiple imputation.
   ```python
   from sklearn.experimental import enable_iterative_imputer
   from sklearn.impute import IterativeImputer
   ```

4. **MissingIndicator**:
   - Generates a binary indicator for missing values, which can be useful for identifying and handling missing data.
   ```python
   from sklearn.impute import MissingIndicator
   ```

These imputers can be used to preprocess your dataset by filling in missing values, which is crucial for many machine learning algorithms that do not handle missing values natively.

````python

age_imputer = SimpleImputer(strategy='mean')
df['age'] = age_imputer.fit_transform(df[['age']])





print(df.isnull().sum())
````

````python
import pandas as pd

# Create the dataset
data = { 
    'date': ['2021-12-01', '01-12-2022', '2022/12/01', '12-01-2021'],
    'country': ['USA', 'U.S.A.', 'America', 'United States'],
    'name': ['John Doe', 'Jonh Doe', 'Jane Doe', 'Jane Doe'],
    'sales_2020': [100, 200, None, 200],
    'sales_2021': [None, 150, 300, 150],
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Convert the 'date' column to a consistent date format using pd.to_datetime()
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Display the DataFrame after date conversion
print("\nDataFrame after date conversion:")
print(df)
````

````python
import pandas as pd
from datetime import datetime

# Create the dataset
data = { 
    'date': ['2021-12-01', '01-12-2022', '2022/12/01', '12-01-2021'],
    'country': ['USA', 'U.S.A.', 'America', 'United States'],
    'name': ['John Doe', 'Jonh Doe', 'Jane Doe', 'Jane Doe'],
    'sales_2020': [100, 200, None, 200],
    'sales_2021': [None, 150, 300, 150],
}

# Convert the dictionary to a DataFrame
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df.head())

# Define a function to parse dates with multiple formats
def parse_date(date_str):
    for i in ('%Y-%m-%d', '%d-%m-%Y', '%Y/%m/%d', '%m-%d-%Y'):
        try:
            return datetime.strptime(date_str, i)
        except ValueError:
            continue
    return None  # Return None if no format matches

# Apply the parse_date function to the 'date' column
df['date'] = df['date'].apply(parse_date)

# Display the DataFrame after date conversion
print("\nDataFrame after date conversion:")
df.head()
````

````python
country_mapping = {'USA':'United States','U.S.A.':'United States','America':'United States'}
df['country']= df['country'].replace(country_mapping)
````

### how do you change the size of the plot

```python
df = sns.load_dataset('titanic')

```

```python
df.info()
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
```

```python
plt.figure(figsize=(20,12)) # change size of the plot
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
```

```python
sns.set(style="whitegrid", rc={'grid.color': '#3C0753'})
plt.figure(figsize=(20, 12),facecolor='#720455')  # Change size of the plot
pallete = {'male':'#030637', 'female':'#910A67'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=pallete)  # Change palette for point color
plt.show()
```
